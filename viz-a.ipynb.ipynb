```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

## Car_Crashes Data Analysis

```python
car_crashes=sns.load_dataset('car_crashes')
```

## Plot alcohol speed RelationShip

```python
sns.relplot(x='speeding', y='alcohol', data=car_crashes, kind='line',color='orange')
```

## Plot Speed total accedent RelationShip

```python
sns.barplot(x='speeding', y='total', data=car_crashes)
```
