```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

# Tips Data Analysis

```python
tips=sns.load_dataset('tips')
```

## Plot Total Bill Versus Tip

```python
sns.lmplot(x='total_bill', y='tip', data=tips, line_kws={'color': 'red'}, scatter_kws={'color': 'black'})
```

## Plot Smoke vs Tip 

```python
sns.boxplot(x='smoker', y='tip', data=tips, color='yellow')
```
